```python
# Name: Chpolpon Zamir,Rhon Vincent Ramos, Nicolas Castiblanco, Brandon Cantey
# Date: 2019-12-25
# Assignment: Final Project

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

"""This Jupyter Notebook will provide insights regarding various video games"""


class VideoGames:
    """
    This class will be providing visualizations regarding platform sales for different regions

    """

    @staticmethod
    def japanplatform():
        """
        This static method will calculate the market share
        of each of the four chosen platforms for Japan


        """

        # get the csv file and read it
        vg_sales_reader = pd.read_csv("vgsales.csv", engine="python")
        # make a dataframe
        vg_df = pd.DataFrame(vg_sales_reader)
        # get the rows
        jps3_rows = vg_sales_reader[vg_sales_reader.Platform == "PS3"]  # get ps3 rows
        jwii_rows = vg_sales_reader[vg_sales_reader.Platform == "Wii"]  # get wii rows
        jx360_rows = vg_sales_reader[vg_sales_reader.Platform == "X360"]  # get xbox rows
        jps2_rows = vg_sales_reader[vg_sales_reader.Platform == "PS2"]  # get ps 2 rows
        # get the market shares for each platform
        jps3_global_sales_avg = (jps3_rows["JP_Sales"].sum() / len(
            jps3_rows["JP_Sales"])) * 1000000  # get the market share number
        jwii_global_sales_avg = (jwii_rows["JP_Sales"].sum() / len(jwii_rows["JP_Sales"])) * 1000000
        jx360_global_sales_avg = (jx360_rows["JP_Sales"].sum() / len(jx360_rows["JP_Sales"])) * 1000000
        jps2_global_sales_avg = (jps2_rows["JP_Sales"].sum() / len(jps2_rows["JP_Sales"])) * 1000000
        # print the dollar amounts
        print("\n\nJapan ps2 sales: $"+str(int(jps2_global_sales_avg)) +
              "00 million")  # print the sales for each platform in millions
        print("Japan ps3 sales: $"+ str(int(jps3_global_sales_avg))+ "00 million")  # print sales for ps 3
        print("Japan wii sales: $"+ str(int(jwii_global_sales_avg))+ "00 million")  # print sales for wii
        print("Japan 360 sales: $"+str(int(jx360_global_sales_avg))+ "00 hundred thousand")  # print sales for xbox360

        # make a list of all the information for the pie
        jpglobal_sales_list = [jps3_global_sales_avg, jx360_global_sales_avg, jwii_global_sales_avg,
                               jps2_global_sales_avg, ]

        labels = "PS3", "XBox360", "Wii", "PS2"  # label the plot
        explode = (0.0, 0.1, 0.0, 0.1)
        colors = ['#FF99CC', '#E0E0E0', '#66B2FF', '#ffcc99', ]  # color the plot
        # make the pie chart
        j_sales_pie = plt.pie(jpglobal_sales_list, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%')

        plt.title(" Japan Sales Average by Platform \n (in millions)")  # title the pie chart
        plt.show()

    @staticmethod
    def europeplatform():
        """

        This is a static method that will calculate each
        of the four chosen platform's market share in Europe

        """
        vg_sales_reader = pd.read_csv("vgsales.csv", engine="python")
        # make a dataframe
        vg_df = pd.DataFrame(vg_sales_reader)

        eps3_rows = vg_sales_reader[vg_sales_reader.Platform == "PS3"]  # get ps 3 rows
        ewii_rows = vg_sales_reader[vg_sales_reader.Platform == "Wii"]  # get wii rows
        ex360_rows = vg_sales_reader[vg_sales_reader.Platform == "X360"]  # get xbox rows
        eps2_rows = vg_sales_reader[vg_sales_reader.Platform == "PS2"]  # get ps 2 rows
        # get the sales for europe
        eps3_global_sales_avg = (eps3_rows["EU_Sales"].sum() / len(
            eps3_rows["EU_Sales"])) * 1000000  # get ps3 europe sales
        ewii_global_sales_avg = (ewii_rows["EU_Sales"].sum() / len(ewii_rows["EU_Sales"])) * 1000000  # wii europe sales
        ex360_global_sales_avg = (ex360_rows["EU_Sales"].sum() / len(
            ex360_rows["EU_Sales"])) * 1000000  # xbox europe sales
        eps2_global_sales_avg = (eps2_rows["EU_Sales"].sum() / len(eps2_rows["EU_Sales"])) * 1000000  # ps2 europe sales
        # print the platform sales in mlns
        print("\n\nEurope ps2 sales: $"+ str(int(eps2_global_sales_avg))+ "0 million")  # ps3
        print("Europe ps3 sales: $"+ str(int( eps3_global_sales_avg))+ "0 million")  # ps3
        print("Europe wii sales: $"+ str(int(ewii_global_sales_avg))+ "0 million")  # wii
        print("Europe 360 sales: $"+ str(int( ex360_global_sales_avg))+ "0 million")  # xbox

        # make europe list of all the information
        eglobal_sales_list = [eps3_global_sales_avg, ex360_global_sales_avg, ewii_global_sales_avg,
                              eps2_global_sales_avg, ]
        # create labels
        labels = "PS3", "XBox360", "Wii", "PS2"

        explode = (0.0, 0.1, 0.0, 0.1)
        # add color
        colors = ['#FF99CC', '#E0E0E0', '#66B2FF', '#ffcc99', ]
        # make a pie chart
        e_sales_pie = plt.pie(eglobal_sales_list, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%')

        plt.title(" Europe Sales Average by Platform \n (in millions)")
        # make sure the plots show in a neat manner
        plt.show()

    @staticmethod
    def northamericaplatform():
        """
        This static method will calculate and visualize the market share of 4 chosen platforms
        in the North America region
        """
        vg_sales_reader = pd.read_csv("vgsales.csv", engine="python")
        # make a dataframe
        vg_df = pd.DataFrame(vg_sales_reader)
        # get the rows
        nps3_rows = vg_sales_reader[vg_sales_reader.Platform == "PS3"]
        nwii_rows = vg_sales_reader[vg_sales_reader.Platform == "Wii"]
        nx360_rows = vg_sales_reader[vg_sales_reader.Platform == "X360"]
        nps2_rows = vg_sales_reader[vg_sales_reader.Platform == "PS2"]
        # get the sales number
        nps3_global_sales_avg = (nps3_rows["NA_Sales"].sum() / len(nps3_rows["NA_Sales"])) * 1000000
        nwii_global_sales_avg = (nwii_rows["NA_Sales"].sum() / len(nwii_rows["NA_Sales"])) * 1000000
        nx360_global_sales_avg = (nx360_rows["NA_Sales"].sum() / len(nx360_rows["NA_Sales"])) * 1000000
        nps2_global_sales_avg = (nps2_rows["NA_Sales"].sum() / len(nps2_rows["NA_Sales"])) * 1000000
        # print the sales in millions
        print("\n\nNorth America ps2 sales: $"+ str(int( nps2_global_sales_avg))+ "0 million")
        print("North America ps3 sales: $"+ str(int( nps3_global_sales_avg))+ "0 million")
        print("North American wii sales: $"+ str(int( nwii_global_sales_avg))+ "0 million")
        print("North American Xbox360 sales: $"+ str(int( nx360_global_sales_avg))+ "0 million")

        # make a list for the pie
        nglobal_sales_list = [nps3_global_sales_avg, nx360_global_sales_avg, nwii_global_sales_avg,
                              nps2_global_sales_avg, ]
        # make labels
        labels = "PS3", "XBox360", "Wii", "PS2"
        explode = (0.0, 0.1, 0.0, 0.1)
        # choose colors
        colors = ['#FF99CC', '#E0E0E0', '#66B2FF', '#ffcc99', ]
        # create the pie
        n_sales_pie = plt.pie(nglobal_sales_list, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%')

        plt.title("  North America Sales Average by Platform \n (in millions) ")
        plt.show()
```

```python
class Genre():
    """
    
    This class will visualize information regarding the video game genres

    """

    @staticmethod
    def get_total_sales_chart_plot():
        """ This static method will visualize the genres
        and their global sales in the form of a bar chart
        """
        # make a set
        vg_sales_reader = pd.read_csv("vgsales.csv", engine="python")
        # make a dataframe
        vg_df = pd.DataFrame(vg_sales_reader)
        genre_set = set()
        genre_names = []
        # for loop to add to a set
        for line in vg_sales_reader["Genre"]:
            genre_set.add(line)  # append, eliminate repeats

        genre_names = ['Action', 'Adventure', 'Fighting', 'Misc', 'Platform', 'Puzzle', 'Racing',
                       'Role-Playing', 'Shooter', 'Simulation', 'Sports', 'Strategy']  # name the genres
        # group by genre
        by_genre = pd.DataFrame(vg_sales_reader.groupby("Genre").sum())  # group by genre
        by_genre_global_sales = by_genre["Global_Sales"]  # genre and global sales
        genre_names = ['Action', 'Adventure', 'Fighting', 'Misc', 'Platform', 'Puzzle', 'Racing',
                       'Role-Playing', 'Shooter', 'Simulation', 'Sports', 'Strategy']  # genre names list
        # make a plot and add colors
        genresales = plt.bar(genre_names, by_genre_global_sales,
                             color=["#7aff73", '#73fff1', '#ff9473', '#ffda73', '#b273ff', '#ff73c0'])
        # title the graph
        plt.title("Global Sale Totals by Genre \n (in millions)")
        # colors = ['#FF99CC','#E0E0E0','#66B2FF','#ffcc99',]

        plt.xticks(rotation=50)
        plt.show()  # appear neat

    @staticmethod
    def get_total_numgames_plot():
        """
        This method will visualize the number of games by genre
        """
        vg_sales_reader = pd.read_csv("vgsales.csv", engine="python")
        # make a dataframe
        vg_df = pd.DataFrame(vg_sales_reader)

        genre_df = vg_df["Genre"].value_counts()  # dataframe with genre counts
        # make a bar plot and add colors
        genre_df_plot = genre_df.plot(kind='bar',
                                      color=["#7aff73", '#73fff1', '#ff9473', '#ffda73', '#b273ff', '#ff73c0'])
        # title the graph
        plt.title("Number of Games by Genre \n (in millions)")
        plt.xticks(rotation=50)
        plt.show()  # neatly show
```

```python
def globalp(publisher, region):
    """This function will print the sales by region according to input
    Args: str publisher, region

    """
    global_only = vg_sales_reader[vg_sales_reader.Publisher == publisher]  # get publisher based on input
    global_only_sales = global_only[region + "_Sales"]  # get the sales for a region based on input
    global_only_sales_sum = global_only_sales.sum()  # get the global sum
    # print,specify mlns
    print("The sum of " + publisher + " sales in " + region + " is $" + str(round(global_only_sales_sum, 2)) + "0.000",
          " mln")
```

```python
def make_gb_plot_graph():
    """
    This function creates a graph of global sales from a range of years
    using a publisher input.
    
    """
    publisher_input = input("Select a publisher/company: ")

    publisher_only = vg_sales_reader[vg_sales_reader.Publisher == publisher_input]
    publisher_sorted = publisher_only.sort_values(['Year'])

    year_only = []
    gb_sales_only = []
    for year in publisher_sorted['Year']:
        year_only.append(year)

    for gb_sale in publisher_sorted['Global_Sales']:
        gb_sales_only.append(gb_sale)

    plt.figure(figsize=(10, 6))
    plt.plot(year_only, gb_sales_only)
    plt.xlabel("Year")
    plt.ylabel("Sales in Millions")
    plt.title("Global Sales of Company from 1980 to 2015")
    plt.show()
```
